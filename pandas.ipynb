```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df1=pd.read_csv('Universities.csv')
```

```python
df1
```

```python
plt.hist(df1['Accept'])
plt.xlabel('Accept')
plt.ylabel('Frequency')
plt.title('Histogram of Accept')
plt.show()
```

```python
plt.hist(df1['SAT'])
plt.xlabel('SAT')
plt.ylabel('Frequency')
plt.title('Histogram of SAT')
plt.show()
```

```python
sn.histplot(df1['Accept'])
plt.show()
```

```python
sn.histplot(df1['SAT'])
plt.show()
```

```python
plt.plot(df1['Accept'],df1['SAT'],'rs')
plt.xlabel('Accept')
plt.ylabel('SAT')
plt.title('Accept vs SAT')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.plot(df1['Accept'],df1['Top10'],'rs')
plt.xlabel('Accept')
plt.ylabel('Top10')
plt.title('Accept vs Top10')
plt.show()
```

```python
sn.histplot(df1)
```
